## Deliverable: Scrape and Analyze Player Data
Create a Beautiful Soup object and use it to scrape the data in the HTML table.

Note that this can also be achieved by using the Pandas `read_html` function. However, use Beautiful Soup here to continue sharpening your web scraping skills.

```python
# Import relevant libraries
from splinter import Browser
from bs4 import BeautifulSoup as soup
import matplotlib.pyplot as plt
import pandas as pd
```

```python
browser = Browser('chrome')
```

```python
#url for Player goals Stats seasons 2017_2019 to 2021_2022
url_1718_player = "https://fbref.com/en/comps/13/2017-2018/stats/2017-2018-Ligue-1-Stats"
url_1819_player = "https://fbref.com/en/comps/13/2018-2019/stats/2018-2019-Ligue-1-Stats"
url_1920_player = "https://fbref.com/en/comps/13/2019-2020/stats/2019-2020-Ligue-1-Stats"
url_2021_player = "https://fbref.com/en/comps/13/2020-2021/stats/2020-2021-Ligue-1-Stats"
url_2122_player = "https://fbref.com/en/comps/13/2021-2022/stats/2021-2022-Ligue-1-Stats"  
```

```python
# Visit the website
urls = [url_1718_player, url_1819_player, url_1920_player, url_2021_player, url_2122_player]
Col = []
dt = []
for url in urls:
    browser.visit(url)
    browser.is_element_present_by_css('div.list_text', wait_time=1)
    html = browser.html
    player_soup = soup(html, 'html.parser')
    table = player_soup.find('table', class_='min_width sortable stats_table shade_zero now_sortable sticky_table eq2 re2 le2')
    # Create an empty list
    columns = []
    data = []
    # Loop through the scraped data to create a list of rows

    headings = table.find_all('th', 'data.stat')
    for heading in headings:
        columns.append(heading.get_data-stat())

    rows = table.find_all('tr')

    for row in rows:
        items = []
        row_data = row.find_all('td')
        for i in row_data:
            items.append(i.text)
        data.append(items)
    Col.append(columns)
    dt.append(data)
    
```
